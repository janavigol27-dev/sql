group by () function : 

```python
data={
    "Department":["IT","HR","Finance","IT","HR","Finance"],
    "Employee":["A","B","C","D","E","F"],
    "Salary":[60000,65000,50000,52000,70000,71000]
}
```

```python
df_hr=pd.DataFrame(data)
```

```python
df_hr
```

```python
df_hr.groupby(by="Department")
```

```python
df_hr.groupby(by="Department")["Salary"]
```

```python
df_hr.groupby("Department")["Salary"].sum()
```

```python
df_hr.groupby("Department")["Salary"].mean()
```

```python
df_hr.groupby(["Department","Employee"])["Salary"].mean()
```

```python
df_hr.groupby("Department")["Salary"].agg(["sum","mean","count"])
```

```python
df_hr.groupby("Department")["Salary"].agg(Total_sum="sum",Avg="mean",Count="count")
```

```python
# Using groupby ().apply with custom function
df=pd.DataFrame({
    "Team":["A","A","B","B","C","C"],
    "Player":["p1","p2","p3","p4","p5","p6"],
    "score":[15,18,10,25,20,23]
})
```

```python
df
```

```python
# df score_diff(x):
# def score_diff(x):
    # return x["Score1"] - x["Score2"]

def score_diff(x):
    return x.max() - x.min()

ans=df.groupby("Team")["score"].apply(score_diff)
ans
```

```python
df
```

```python
# Fitter groups with team total score > 35
total_score=df.groupby("Team")["score"].sum()
high_score_teams=total_score[total_score>35].index
filtered_df=df[df["Team"].isin(high_score_teams)]
filtered_df
```
