Mearge data & join data.

```python
# df1 = {
#     "ID":[1,2,3],
#     "name":["alice","bob","charlie"],
#     "age":[35,22,45]
# }
```

```python
df1_dict = {
    "ID":[1,2,3],
    "name":["alice","bob","charlie"],
    "age":[35,22,45]
}

df2_dict = {
    "ID":[3,4],
    "salary":[45000,35000]
}

# Convert to DataFrame
df1 = pd.DataFrame(df1_dict)
df2 = pd.DataFrame(df2_dict)

# Merge
result = pd.merge(df1, df2, on="ID", how="inner")
print(result)
```
